### Об игре
---

AI Dungeon 2 -это бесплатная однопользовательская текстовая приключенческая игра, которая использует искусственный интеллект для генерации неограниченного контента.

Русская версия игры сделана на основе [Clover-Edition](https://github.com/cloveranon/Clover-Edition) - [MIT](https://github.com/cloveranon/Clover-Edition/blob/master/LICENSE) и русской версии GPT-3

За более подробными инструкциями и советами можно воспользоваться английской [Wiki](https://github.com/VBPXKSMI/Open-CYOAI-Project/wiki).

---
### Установка
---

Для более удобного использования рекомендуется установить темную тему. ("Tools"-> "Settings..." -> "Theme" -> "Dark"*)

Для игры требуется:
1. Google drive с минимум 6гб свободного места.
2. В меню выберите "File"-> "Save a a copy in Drive". Закройте данное окно и начните играть в новом окне.

---
### Как играть
---

1. (Опционально) Перейдите в секцию отрисовка рендера, чтобы изменить настройки отображения текста игры.
2. Перейдите в секцию НАЧАТЬ ИГРУ, выберите одну из предложенных моделей (сейчас только Clover-Russian-edition)
